```python
import matplotlib.pyplot as plt
```

```python
x = ['python','c','c++','java']
y = [90,20,50,30]

plt.xlabel('Language')
plt.ylabel('Demand')
plt.title('Language in Demand',fontsize =15)

c = ['r','b','g','m']
plt.bar(x, y, width=0.5, color =c, label ='Popluarity')
plt.legend()
plt.show()
```

```python
x = ['python','c','c++','java']
y = [90,20,50,30]

plt.xlabel('Demand')
plt.ylabel('Language')
plt.title('Language in Demand',fontsize =15)

c = ['r','b','g','m']
plt.barh(x, y, height=0.5, color =c, label ='Popluarity')
plt.legend()
plt.show()
```
