```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [10,20,30,15,40,10,60,35]

plt.scatter(x, y, c='r', s=100, alpha=0.5, marker='*')

plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day')
plt.ylabel('Price')

plt.show()
```

```python
x1 = [1,2,3,4,5,6,7,8]
y1 = [10,20,30,15,40,10,60,35]

x2 = [1,2,3,4,5,6,7,8]
y2 = [20,10,40,25,50,30,5,55]

plt.scatter(x1, y1, c='r', s=100, alpha=0.5)
plt.scatter(x2, y2, c='g', s=50, alpha=0.7)

plt.legend(["r", "g"], loc ="upper left")
plt.title('Scatter Plot', fontsize=15)
plt.xlabel('Day')
plt.ylabel('Price')

plt.show()
```
