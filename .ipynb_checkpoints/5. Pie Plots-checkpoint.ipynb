```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python']

ex = [0.0,0.0,0.0,0.2]

plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", shadow=True, radius=2)

plt.show()
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python']

ex = [0.0,0.0,0.0,0.2]

plt.pie(x, labels=y, explode=ex, shadow=True, radius=2, labeldistance=1.1, startangle=90)

plt.show()
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python']

ex = [0.0,0.0,0.0,0.2]

plt.pie(x, labels=y, explode=ex, shadow=True, radius=2, autopct='%0.1f%%',
       textprops={'fontsize': 15})

plt.legend(loc='lower left')
plt.show()
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python']

ex = [0.0,0.0,0.0,0.2]

plt.pie(x, labels=y, explode=ex, shadow=True, radius=2,
       textprops={'fontsize': 15})
plt.pie([1], colors='w')

plt.legend(loc='lower left')
plt.show()
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python']

ex = [0.0,0.0,0.0,0.2]

plt.pie(x, labels=y, explode=ex, shadow=True, radius=2, autopct='%0.1f%%',
       textprops={'fontsize': 15})
cr = plt.Circle(xy=(0,0), radius=0.5, facecolor='w')
plt.gca().add_artist(cr)

plt.legend(loc='lower left')
plt.show()
```
