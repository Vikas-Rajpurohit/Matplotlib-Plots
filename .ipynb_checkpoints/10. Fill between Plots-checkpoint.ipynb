```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,5,3,4]
y = [1,2,6,2,3]

plt.plot(x,y,color='g', marker='*', markersize=10)

plt.fill_between(x,y)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,2,3,6]

plt.plot(x,y,color='y', marker='*', markersize=10)

plt.fill_between(x=[1.5,2.5],y1=1,y2=3)

plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,2,3,6])

plt.plot(x,y,color='g', marker='*', markersize=10)

plt.fill_between(x,y, where=(x>=2) & (x<=4))

plt.show()
```
