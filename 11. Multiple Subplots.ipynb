```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

y = [y1,y2,y3]

num_subplots = 3
rows = 2
cols = (num_subplots + rows - 1) // rows 

fig, axs = plt.subplots(rows, cols, figsize=(10, 6))

# Flatten the axs array if needed
if num_subplots > 1:
    axs = axs.flatten()

# Loop through the subplots and plot the data
for i in range(num_subplots):
    ax = axs[i]
    ax.plot(x, y[i])
    ax.set_title(f'Plot {i+1}')

# Adjust the layout and spacing
plt.tight_layout()

# Show the plots
plt.show()
```
