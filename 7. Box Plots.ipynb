```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert=False, labels=['Marks'])

plt.show()
```

```python
x = [10,20,30,40,50,60,120,200]

plt.boxplot(x,whis=1,showmeans=True)

plt.show()
```

```python
x1 = [10,20,30,40,50,80,120]
x2 = [30,40,50,70,100,150]

y = [x1,x2]

plt.boxplot(y, whis=1, labels=['Marks','Score'], showmeans=True)

plt.show()
```
